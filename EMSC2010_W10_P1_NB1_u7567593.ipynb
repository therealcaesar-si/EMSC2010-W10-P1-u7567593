In this notebook we'll calculate and plot some sinusoids and see how you can change their frequency, amplitude, and phase.

We'll also mix cycles together and see how you quickly obtain a complicated signal.

Start by importing the packages we'll need.

````python
import numpy as np #numpy for numerical operations
import matplotlib.pyplot as plt #matplotlib for plotting
````

First we'll define times between 0 and 800 with a spacing of 0.1, then we'll calculate a sinusoid with a frequency of 0.01 (i.e. a period of 100).

````python
time = np.arange(0,800,0.1) #generate times with a spacing of 0.1 between 0 and 799
f = 0.01 #frequency of the sinusoid
# period is therefore 100, f = 1/period = 1/100 = 0.01
signal = np.sin(2*np.pi*f*time) #calculate the signal
plt.plot(time,signal) #plot the signal
plt.xlim([0,800]) #limit the x-axis
plt.minorticks_on() #add minorticks
plt.xlabel('Time') #label the x-axis
plt.ylabel('Signal') #label the y-axis
````

It is simple to change the frequency, just be changing ```f```.

In the code below we will make a sinusoid with a frequency of 0.02 (period of 50).

````python
time = np.arange(0,800,0.1) #generate times with a spacing of 0.1 between 0 and 799
f = 0.02 #frequency of the sinusoid
signal = np.sin(2*np.pi*f*time) #calculate the signal
plt.plot(time,signal) #plot the signal
plt.xlim([0,800]) #limit the x-axis
plt.minorticks_on() #add minorticks
plt.xlabel('Time') #label the x-axis
plt.ylabel('Signal') #label the y-axis
````

We can adjust the amplitude of the sinusoid by multiplying the signal by the desired amplitude.

In the example below we'll use an amplitude of 2.5.

````python
A = 2.5 #amplitude of the sinusoid
f = 0.01 #frequency of the sinusoid
signal = np.sin(2*np.pi*f*time)*A #calculate the signal
plt.plot(time,signal) #plot the signal
plt.xlim([0,800]) #limit the x-axis
plt.minorticks_on() #add minorticks
plt.xlabel('Time') #label the x-axis
plt.ylabel('Signal') #label the y-axis
````

To adjust the phase we need to add an angle (in radians) into the term within the ```np.sin()``` function.

In this case we'll change the phase to $\pi$/2, which is equivilent to a quarter of a cycle (where a full cycle is 2$\pi$).

````python
A = 2.5 #amplitude of the sinusoid
f = 0.01 #frequency of the sinusoid
p = np.pi/2 #phase of the sinusoid
signal = np.sin(2*np.pi*f*time+p)*A #calculate the signal
plt.plot(time,signal) #plot the signal
plt.xlim([0,800]) #limit the x-axis
plt.minorticks_on() #add minorticks
plt.xlabel('Time') #label the x-axis
plt.ylabel('Signal') #label the y-axis
````

We can combine sinusoids simply by adding them together. In this example we'll create three sinusoids with:

1.   Frequency = 1/100, Amplitude = 2
2.   Frequency = 1/40, Amplitude = 1
3.   Frequency = 1/20, Amplitude = 0.5

We'll calculate the sinusoids separately and then add them together.


````python
f1 = 1/100 #frequency of the 1st signal
f2 = 1/40 #frequency of the 2nd signal
f3 = 1/20 #frequency of the 3rd signal

A1 = 2 #amplitude of the 1st signal
A2 = 1 #amplitude of the 2nd signal
A3 = 0.5 #amplitude of the 3rd signal

signal1 = np.sin(2*np.pi*f1*time)*A1 #first signal
signal2 = np.sin(2*np.pi*f2*time)*A2 #second signal
signal3 = np.sin(2*np.pi*f3*time)*A3 #third signal
signal = signal1 + signal2 + signal3 #combine the signals

plt.plot(time,signal) #plot the signal
plt.xlim([0,800]) #limit the x-axis
plt.minorticks_on() #add minorticks
plt.xlabel('Time') #label the x-axis
plt.ylabel('Signal') #label the y-axis
````
